Analysis for Proposition 062 - Death Penalty. Initiative Statute.
Where we look at who the major donors are for and against the proposition to repeal the death penalty.

```python
import pandas as pd
```

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head(7)
```

```python
props.info()
```

```python
contris = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contris.info()
```

```python
contris.head()
```

```python
props.prop_name.value_counts()
```

```python
prop64 = props[(props.prop_name == "PROPOSITION 062- DEATH PENALTY. INITIATIVE STATUTE.")]
```

```python
prop64.info()
```

```python
contris.info()
```

```python
merged = pd.merge(prop64, contris, on= "calaccess_committee_id")
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / merged.amount.sum()
```

```python
oppose.sort_values("amount", ascending=False).head(10)
```

```python
merged.groupby(["contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
